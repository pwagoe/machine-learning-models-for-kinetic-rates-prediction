```python
#define path and import libraries
import os
import pandas as pd
import numpy as np

os.chdir("C:/Users/Pauli/OneDrive/Dokumente/Uni/Masterarbeit/code/preprocessing")

#load data from current directory
paper_data = pd.read_excel("ja8b08048_si_002.xlsx")
#delete rows 0-6
paper_data = paper_data.drop(index=paper_data.index[0:8])
# rename the right column for k_off
paper_data.rename(columns={paper_data.columns[13]: 'k_off'}, inplace=True)
paper_data.rename(columns={paper_data.columns[0]: 'kinase'}, inplace=True)
paper_data.rename(columns={paper_data.columns[1]: 'ligand'}, inplace=True)
# keep only desired columns
df_clean = paper_data[['kinase', 'ligand', 'k_off']]
print(f"erstmal alle daten: {df_clean.shape}")
#drop rows with missing k_off values
df_clean = df_clean.dropna(subset=['k_off'])
print(f"leere dateneinträge für koff weg: {df_clean.shape}")
#convert k_off to numeric
df_clean['k_off'] = pd.to_numeric(df_clean['k_off'], errors='coerce')
#drop rows with semi-quantitative k_off values = 0.35
df_clean = df_clean[df_clean['k_off'] != 0.35]
print(f"semiquantitative einträge weg: {df_clean.shape}")
#reset index
df_clean = df_clean.reset_index(drop=True)
#log10 transform k_off values and add as new column
df_clean["log_koff"] = np.log10(df_clean["k_off"].values).reshape(-1, 1)
print(f"final: {df_clean.shape}")
df_clean
```
